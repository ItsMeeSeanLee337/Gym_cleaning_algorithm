```python
#Rearranging columns
#Business type first column
#Clean name second column
#All names should be only alphanumeric
df = df[['First Name'] + [col for col in df.columns if col != 'First Name']]
df = df[['Business Type'] + [col for col in df.columns if col != 'Business Type']]
df
```

```python
#All names should be only alphanumeric
df['First Name'] = df['First Name'].str.replace("[^A-Za-z0-9 &']", "", regex=True)
df
```

```python
#Remove city from the name
#First name should not have the city name, name should have the city name in parenthesis
def findLocationFirstName(First_Name, search_city):
    if (search_city in First_Name):
        First_Name = First_Name.replace(search_city, '')
        return First_Name
    else:
        return First_Name
def findLocationName(name, search_city):
    if (search_city in name):
        name = name.replace(search_city, '(' + search_city + ')')
        return name
    else:
        return name
df['First Name'] = df.apply(lambda row: findLocationFirstName(row['First Name'], row['search_city']), axis = 1)
df['name'] = df.apply(lambda row: findLocationName(row['name'], row['search_city']), axis = 1)
df
```
